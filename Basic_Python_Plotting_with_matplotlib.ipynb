#Handling Date Values

```
import datetime
events = [datetime.date(2015,1,23),datetime.date(2015,1,28),
          datetime.date(2015,2,3),datetime.date(2015,2,21),
          datetime.date(2015,3,15),datetime.date(2015,3,24),
          datetime.date(2015,4,8),datetime.date(2015,4,24)]
readings = [12,22,25,20,18,15,17,14]
plt.plot(events, readings)
```

**matplotlib.dates** module specialized for managing dates

**MonthLocator()** and **DayLocator()** define the scales of the months and the days, respectively

**DateFormatter()** formats the dates in your own preference

Once having defined the scales, use **set_major_locator()** and **set_minor_locator()** on the **xaxis** object and use **set_major_formatter()** to set the text format of the tick labels

```
import matplotlib.dates as mdates
months = mdates.MonthLocator()
days = mdates.DayLocator()
timeFmt = mdates.DateFormatter('%Y-%m')

fig, ax = plt.subplots() #get figure object and axes object
plt.plot(events, readings)

ax.xaxis.set_major_locator(months)
ax.xaxis.set_minor_locator(days)
ax.xaxis.set_major_formatter(timeFmt)
```

Suppose you want to plot,

```
x = np.arange(-2*np.pi, 2*np.pi, 0.01)
y = np.sin(3*x)/x
plt.plot(x,y)
```

To display a family of functions,

$y = sin(n*x)/x$

```
x = np.arange(-2*np.pi,2*np.pi,0.01)
y = np.sin(x)/x
y2 = np.sin(2*x)/x
y3 = np.sin(3*x)/x
plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
```

```
plt.plot(x,y,'k--',linewidth=2)
plt.plot(x,y2,'m-.')
plt.plot(x,y3,color='#87a3cc',linestyle='--')
```

Since the x range is from -2$\pi$ to 2$\pi$, it is more reasonable to use multiple of $\pi$ as the values shown on x-axis.

Use the **xticks()** and **yticks()** passing to each of them two lists of values,

first list contains values corresponding to the positions where the ticks are to be placed,

second list contains the tick labels in string...

Since you have to display $\pi$, you should use LaTeX format staring with r prefix...enclosed by two $'s.'

```
plt.plot(x,y,color='b')
plt.plot(x,y2,color='r')
plt.plot(x,y3,color='g')

'''Change x ticks'''
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
           [r'$-2\pi$',r'$-\pi$',r'$0$',r'$+\pi$',r'$+2\pi$'])
```

To have the x-axis and y-axis passing through the origin to display graph,

1.  you must capture the Axes object through the **gca()** function.

2.  Then set the four sides of the bounding box specifying for each one its position: right, left, top, bottom. Crop the sides by using the **set_color()** under the **gca().spines['side']** and indicating **None** for color.

3.  Then Use **set_position()** to move the x- and y-axes to the origin.



```
plt.plot(x,y,color='b')
plt.plot(x,y2,color='r')
plt.plot(x,y3,color='g')
'''set xticks and yticks'''
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
           [r'$-2\pi$',r'$-\pi$',r'$0$',r'$+\pi$',r'$+2\pi$'])
plt.yticks([-1,0,+1,+2,+3],
           [r'$-1$',r'$0$',r'$+1$',r'$+2$',r'$+3$'])

#step 1
ax = plt.gca()

#step 2
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

#step 3
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))
```

To specify a particular point of the line using a notation and optionally add an arrow to better indicate the position of the point, use **annotate()**

###annotate()###
first argument = string expression in LaTex,

**xy** kwarg = position

**xytext** kwarg = distance of textual notation from the point of interest

**arrowprops** = =arrow properties

```
plt.plot(x,y,color='b')
plt.plot(x,y2,color='r')
plt.plot(x,y3,color='g')
plt.xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi],
           [r'$-2\pi$',r'$-\pi$',r'$0$',r'$+\pi$',r'$+2\pi$'])
plt.yticks([-1,0,+1,+2,+3],
           [r'$-1$',r'$0$',r'$+1$',r'$+2$',r'$+3$'])
ax = plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

#Example: annotate()
plt.annotate(r'$\lim_{x\to 0}\frac{\sin(x)}{x}= 1$', #math expression
             xy=[0,1],
             xycoords='data',
             xytext=[30,30],
             fontsize=16,
             textcoords='offset points',
             arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.2")
            )
```

##Bar Charts
**bar()** function

```
index = np.array([0,1,2,3,4])
values = [5,7,3,4,6]
plt.bar(index, values)
```

The indexes drawn on the x-axis corresponds to a category. You can specify the categories through the tick label, defined by a list of strings passed to the **xticks()** function.

The location of these tick labels are specified in the first argument of the **xticks()** function.

```
plt.subplot(121)
plt.bar(index, values)
plt.xticks(index, ['A','B','C','D','E'])
plt.subplot(122)
plt.bar(index, values)
plt.xticks(index+0.4, ['A','B','C','D','E'])
```

Useful kwargs for **bar()**:
*  **yerr** - a list containing the standard deviations
*  **error_kw** contains other kwargs:
  *  **eColor** - color of the error bars,
  *  **capsize** - width of the transverse lines marking the ends of the error bars
  *  **alpha** - degree of transparency from 0 to 1
  *  **label** - identify the series presented

```
index = np.arange(5)
values1 = [5,7,3,4,6]
std1 = [0.8,1,0.4,0.9,1.3]
plt.title('A Bar Chart')

#refine bar()
plt.bar(index, values1,
        yerr=std1,
        error_kw={'ecolor':'0.1',
                  'capsize':50,
                  'alpha': 0.7,
                  'label': 'First'})
plt.xticks(index+0.4, ['A','B','C','D','E'])
plt.legend(loc=2)
```

###Horizontal Bar Charts
**barh()** function

kwargs and arguments in **bar()** still valid for **barh()**

```
plt.title('A Horizontal Bar Chart')
plt.barh(index, values1,
         xerr=std1,
         error_kw={'ecolor':'0.01',
                   'capsoze':6,
                   'alpha':0.7,
                   'label':'First'})
plt.yticks(index+0.4, ['A','B','C','D','E'])
plt.legend(loc=5)
```

###Multiseries Stacked Bar Charts
add **bottom** kwarg to each **bar()** function

```
series1 = np.array([3,4,5,3])
series2 = np.array([1,2,2,5])
series3 = np.array([2,3,3,4])
index = np.arange(4)
plt.axis([-0.5,3.5,0,15])
plt.title('A Multiseries Stacked Bar Chart')

#don't call bottom in the first plot
plt.bar(index, series1, color='r')
#add bottom on top of series1
plt.bar(index, series2, color='b', bottom=series1)
#add bottom on top of series2
plt.bar(index, series3, color='g', bottom=(series2+series1))

plt.xticks(index, ['Jan18','Feb18','Mar18','Apr18'])
```

For the horizontal stacked bar chart, the **bottom** kwarg should be replaced by the **left** kwarg

```
plt.title('A Multiseries Horizontal Stacked Bar Chart')
#use left instead of bottom kwargs
plt.barh(index,series1,color='r')
plt.barh(index,series2,color='g',left=series1)
plt.barh(index,series3,color='b',left=(series1+series2))
plt.yticks(index,['Jan18','Feb18','Mar18','Apr18'])
```

You can also use hatches instead of solid colors in bar charts. To do this, you have to first set the **color** of the bar as white, and then use **hatch** kwarg to define the hatches.

Hatch characters: |, /, -, \, *, x

The more symbols replicated, the denser the lines forming the hatch.

```
plt.axis([0,15,-0.5,3.5])
plt.title('A Multiseries Horizontal Stacked Bar Chart')
#set color to white
plt.barh(index,series1,color='w',hatch='xx')
plt.barh(index,series2,color='w',hatch='///', left=series1)
plt.barh(index,series3,color='w',hatch='\\\\\\',left=(series1+series2))
plt.yticks(index+0.4,['Jan18','Feb18','Mar18','Apr18'])
```

##Pie Charts
**pie()** function

You need to define key features making use of the kwargs as a list.

```
#Define a list of labels
labels = ['Nokia','Samsung','Apple','Lumia']
#Define values, don't have to be a sum of 100
values = [15, 55, 75, 5]
#Define plotting colors
colors = ['yellow','green','red','blue']

#axis() not set to equal
plt.subplot(121)
plt.pie(values, labels=labels, colors=colors)
#axis() set to equal
plt.subplot(122)
plt.pie(values, labels=labels, colors=colors)
plt.axis('equal')
```

##Advanced Charts

###Contour Plots
need to define $z=f(x,y)$ for generating a three-dimensional surface

**contour()** function

```
dx = 0.01; dy = 0.01
x = np.arange(-2.0, 2.0, dx)
y = np.arange(-2.0, 2.0, dy)
#create a mesh for contour
X,Y = np.meshgrid(x,y)

#define a third dimension,
def f(x,y):
  return (1 - y**5 + x**5)*np.exp(-x**2 - y**2)

#define color gradient levels
color_levels = 8
#plot contour lines
C = plt.contour(X,Y,f(X,Y), color_levels, colors='red')
#plot contour filled colors
plt.contourf(X,Y,f(X,Y), color_levels)
#plot contour label value
plt.clabel(C, inline=1, fontsize=15)
```

**cmap** kwarg specify the color maps filled in the **contourf()** function

**colorbar()** adds a color scale

```
#define color gradient levels
color_levels = 10
#plot contour lines
C = plt.contour(X,Y,f(X,Y), color_levels, colors='black')
#plot contour filled colors
plt.contourf(X,Y,f(X,Y), color_levels, cmap=plt.cm.hot)
#                                       ^specify a colormap
#plot contour label value
plt.clabel(C, inline=1, fontsize=15)
#add a colorbar
plt.colorbar()
```

###3D Surfaces
In the previous example, you used **contour()** to represent a three-dimensional surface. Now you can use mplot3D package to plot a 3D plot.

**plot_surface() function

```
#create a figure object
fig = plt.figure()
#create a axes object
ax = Axes3D(fig)

X = np.arange(-2,2,0.1)
Y = np.arange(-2,2,0.1)
X,Y = np.meshgrid(X,Y)
def f(x,y):
  return (1 - y**5 + x**5)*np.exp(-x**2-y**2)

#plot
ax.plot_surface(X,Y,f(X,Y), rstride=1, cstride=1)
```

**cmap** in **plot_surface()** function changes the colormap

**view_init()** function changes the view point:
* **elev** adjusts the height at which the surface is seen,
* **azim** adjusts the angle of rotation of the surface

```
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-2,2,0.1)
Y = np.arange(-2,2,0.1)
X,Y = np.meshgrid(X,Y)
def f(x,y):
  return (1 - y**5 + x**5)*np.exp(-x**2-y**2)

#plot
ax.plot_surface(X,Y,f(X,Y), rstride=1, cstride=1,
                cmap=plt.cm.hot) #colormap
ax.view_init(elev=30, azim=125)
```

###Grids of Subplots
**GridSpec()** function allows splitting the drawing area into a grid of sub-areas, to which you can assgin one or more of them to each subplot, so that in the end you can obtain subplots with different sizes and orientations.

```
gs = plt.GridSpec(3,3) #create a 3x3 grid
fig = plt.figure(figsize=(6,6)) #figsize = a tuple of intergers: width, height in inches
fig.add_subplot(gs[1,:2]) #This subplot takes a grid in row 1, and columns 0,1
fig.add_subplot(gs[0,:2]) # row 0, columns 0,1
fig.add_subplot(gs[2,0])  # row 2, column 0
fig.add_subplot(gs[:2,2]) # rows 0,1, column 2
fig.add_subplot(gs[2,1:]) # row 2, columns 1,2
```

You can use this Axes object returned by each **add_subplot()** function to call the **plot()** function to draw the corresponding plot

```
x1 = np.array([1,3,2,5])
y1 = np.array([4,3,7,2])
x2 = np.arange(5)
y2 = np.array([3,2,4,6,4])

gs = plt.GridSpec(3,3)
fig = plt.figure(figsize=(6,6))

#assign a variable to the add_subplot function to call plot
s1 = fig.add_subplot(gs[1,:2])
s1.plot(x1,y1,'r')

s2 = fig.add_subplot(gs[0,:2])
s2.bar(x2,y2)

s3 = fig.add_subplot(gs[2,0])
s3.barh(x2,y2, color='g')

s4 = fig.add_subplot(gs[:2,2])
s4.plot(x2,y2,'k')

s5 = fig.add_subplot(gs[2,1:])
s5.plot(x1,y1,'b^',x2,y2,'yo')
```
